# Lectura de datos ADC desde ESP32

Captura de señales analógicas mediante el convertidor analógico-digital del ESP32 a través de comunicación serial.

## 1. Importación de bibliotecas

```python
import serial
import time
import csv
import plotly.graph_objects as go
import numpy as np
from datetime import datetime
```

## 2. Configuración de parámetros

```python
PUERTO_SERIAL = '/dev/ttyUSB0' #Depende de tu puerto y SO
BAUDRATE = 115200 # baudios
TIEMPO_LECTURA = 5.0 # segundos
ARCHIVO_CSV = f'datos_adc_{datetime.now().strftime("%Y%m%d_%H%M%S")}.csv'
```

## 3. Adquisición de datos

```python
datos = []

ser = serial.Serial(PUERTO_SERIAL, BAUDRATE, timeout=1)
time.sleep(2)
ser.reset_input_buffer()

tiempo_inicio = time.time()
while (time.time() - tiempo_inicio) < TIEMPO_LECTURA:
    if ser.in_waiting > 0:
        linea = ser.readline().decode('utf-8').strip()
        try:
            datos.append(int(linea))
        except:
            print()(f"Valor no numérico recibido: {linea}")

ser.close()
print(f"Muestras adquiridas: {len(datos)}")
```

## 4. Exportación a CSV

```python
with open(ARCHIVO_CSV, 'w', newline='') as archivo:
    escritor = csv.writer(archivo)
    escritor.writerow(['Valor_ADC'])
    for valor in datos:
        escritor.writerow([valor])

print(f"Archivo generado: {ARCHIVO_CSV}")
```

## 5. Visualización

```python
fig = go.Figure()
fig.add_trace(go.Scatter(y=datos, mode='lines', name='Señal ADC'))
fig.update_layout(
    xaxis_title='Índice de muestra',
    yaxis_title='Valor ADC',
    template='plotly_white'
)
fig.show()
```

## 6. Estadísticas

```python
datos_np = np.array(datos)
print(f"Muestras: {len(datos)}")
print(f"Media: {np.mean(datos_np):.2f}")
print(f"Desviación estándar: {np.std(datos_np):.2f}")
print(f"Mínimo: {np.min(datos_np)}")
print(f"Máximo: {np.max(datos_np)}")
```
